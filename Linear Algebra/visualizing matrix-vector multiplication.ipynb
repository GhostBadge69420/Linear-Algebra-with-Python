M = [.5 1

v = [1

     a]

a ∈ {-2,...,2}

```python
vcomp = np.linspace(-2,2,40)

for a in vcomp:
    v = np.array([1,a])
    Mv = M@v
    plt.plot([0,v[0]],[0,v[1]])
    plt.plot([0,Mv[0]],[0,Mv[1]])

plt.show()
```

```python
vcomp = np.linspace(-2,2,40)

for a in vcomp:
    v = np.array([1,a])
    Mv = M@v
    plt.plot([0,v[0]],[0,v[1]],color=[1-abs(a)/4,.5,abs(a)/2],alpha=.8)
    plt.plot([0,Mv[0]],[0,Mv[1]],color=[1-abs(a)/2,abs(a)/4,.5])

plt.axis('square')
plt.axis('off')
plt.show()
```
