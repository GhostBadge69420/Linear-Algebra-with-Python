```python
v = np.array([.5,1])
s = [1,-.5,2,.5]

for si in s:
    sv = si*v
    plt.plot([0,sv[0]],[0,sv[1]],'o-',linewidth=3,label='$\\lambda=%g$'%si)

plt.show()
```

```python
v = np.array([.5,1])
s = [1,-.5,2,.5]

for si in s:
    sv = si*v
    plt.plot([0,sv[0]],[0,sv[1]],'o-',linewidth=3,label='$\\lambda=%g$'%si)

plt.axis('square')
plt.axis([-3,3,-3,3])
plt.grid()
plt.legend()
plt.show()
```

```python
v1 = np.array([-1,2])
v2 = np.array([1,1])

v3a = v1+v2
v3b = np.add(v1,v2)
v3c = np.zeros(2)
for i in range(0,2):
    v3c[i] = v1[i] + v2[i]

print(v3a,v3b,v3c)
```
