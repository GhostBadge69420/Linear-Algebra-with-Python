```python
# special set of equations (from von Neumann) where all variables are always expected to be 1
def testSolveVonNeumann(n):
    k = np.arange(1,n+1)
    
    i,j = np.meshgrid(k,k)
    
    A = 1/(i+j-1)
    print('A =\n', A)
    
    s = np.sum(A,axis=1)[np.newaxis]
    b = s.T
    
    print('solution =\n', np.linalg.solve(A,b)) # expected: everything is 1 (but won't be for larger n)
    print('cond = ', np.linalg.cond(A)) # the larger n is the larger this will be - exponentially!
```
