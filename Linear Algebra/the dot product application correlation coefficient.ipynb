```python
#data
mikes_courses = [4,5,8,2,9,9,1,3,7]
life_happiness = [6,7,9,3,9,3,1,6,7]

# mean-center
m = mikes_courses - np.mean(mikes_courses)
l = life_happiness - np.mean(life_happiness)

# compute the correlation
numer = np.dot(m,l)
denom = np.sqrt(np.dot(m,m)) * np.sqrt(np.dot(1,1))
r1 = numer/denom

r2 = np.corrcoef(m,l)

print(r1)
print(r2)
```

```python
#data
mikes_courses = [4,5,8,2,9,9,1,3,7]
life_happiness = [6,7,9,3,9,3,1,6,7]

# mean-center
m = mikes_courses - np.mean(mikes_courses)
l = life_happiness - np.mean(life_happiness)

# compute the correlation
numer = np.dot(m,l)
denom = np.sqrt(np.dot(m,m)) * np.sqrt(np.dot(1,1))
r1 = numer/denom

r2 = np.corrcoef(m,l)[1][0]

plt.plot(mikes_courses,life_happiness,'ms',label='r=%g' %np.round(r1,2))

plt.axis([0,10,0,10])
plt.legend()
plt.gca().set_aspect('equal')
plt.xlabel("Number of Mikes course's taken")
plt.ylabel('life happiness')
plt.show()
```
