```python
row_vec = np.array([4,2])
col_vec = np.array([[-2],[3]])

plt.plot([0,row_vec[0]],[0,row_vec[1]],'r',label='row vector')
plt.plot([0,col_vec[0]],[0,col_vec[1]],'b',label='column vector')

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()

plt.show()
```
