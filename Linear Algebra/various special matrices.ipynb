```python
inmat = np.random.randint(-4,5,(4,4))
print(inmat)
```

```python
inmat = np.random.randint(-4,5,(4,4))
I4 = np.eye(4)
upper4 = np.triu(np.random.randn(4,4))

print(inmat)
print(inmat@I4)
```

```python
inmat = np.random.randint(-4,5,(4,4))
I4 = np.eye(4)
upper4 = np.triu(np.random.randn(4,4))

print(inmat)
print('\n')
print(inmat@I4)
```

```python
inmat = np.random.randint(-4,5,(4,4))
I4 = np.eye(4)
upper4 = np.triu(np.random.randn(4,4))

print(type(inmat))
print('\n')
print(type(inmat@I4))
```

```python
inmat = np.random.randint(-4,5,(4,4))
I4 = np.eye(4)
upper4 = np.triu(np.random.randn(4,4))

print(inmat.dtype)
print('\n')
print((inmat@I4).dtype)
```
