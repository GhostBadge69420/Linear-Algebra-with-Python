```python
U,s,V = np.linalg.svd(ein)

fig,ax = plt.subplots(1,3)

ax[0].imshow(U)
ax[0].set_title('U')

ax[1].imshow(np.diag(s))
ax[1].set_title('$\\Sigma$')

ax[2].imshow(V)
ax[2].set_title('V')

plt.show()
```

# Exercise 1

Bring Einstein back from dead!

```python
S = np.zeros(np.shape(ein))
for i in range(0,len(s)):
    S[i,i] = s[i]

rein = U@S@V

plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('original')

plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD version')
```
