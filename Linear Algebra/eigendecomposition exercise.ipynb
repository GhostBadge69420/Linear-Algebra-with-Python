1) Extract eigenvectors and eignevalues from M = [1 2
                      
                                                  2 1]

2) Plot Î»v and Mv


```python
M = np.array([[1,2],[2,1]])
d,V = np.linalg.eig(M)

print(d)
```

```python
M = np.array([[1,2],[2,1]])
d,V = np.linalg.eig(M)

lv = d[0]*V[:,0]
Mv = M@V[:,0]

plt.plot([0,lv[0]],[0,lv[1]],'b',label='$\\lambda v$',linewidth=4)
plt.plot([0,Mv[0]],[0,Mv[1]],'r:',label='Mv',linewidth=4)

plt.legend()
plt.grid()
plt.show()
```

```python
M = np.array([[1,2],[2,1]])
d,V = np.linalg.eig(M)

lv = d[0]*V[:,0]
Mv = M@V[:,0]

plt.plot([0,lv[0]],[0,lv[1]],'b',label='$\\lambda v$',linewidth=4)
plt.plot([0,Mv[0]],[0,Mv[1]],'r:',label='Mv',linewidth=4)

plt.axis('square')
plt.axis([-3,3,-3,3])
plt.legend()
plt.grid()
plt.show()
```

```python
m = 6
Q = np.random.randn(m,m)
Q = Q@Q.T 

d,V = np.linalg.eig(Q)

print(V)
```

```python
m = 6
Q = np.random.randn(m,m)
Q = Q@Q.T 

d,V = np.linalg.eig(Q)

dps = np.zeros((m,m))

for i in range(m):
    for j in range(0,m):
        dps[i,j] = np.dot(V[:,i],V[:,j])

print(dps)
```

```python
m = 6
Q = np.random.randn(m,m)
Q = Q@Q.T 

d,V = np.linalg.eig(Q)

dps = np.zeros((m,m))

for i in range(m):
    for j in range(0,m):
        dps[i,j] = np.dot(V[:,i],V[:,j])

print(np.round(dps,2))
```
